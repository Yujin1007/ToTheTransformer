## 딥러닝을 가능케한 autograd

```python
x=torch.tensor([1.],requires_grad=True) # float 이여야해서 1.
print(x)
```

```python
x=torch.tensor([1.])
print(x)
print(x.requires_grad)

x.requires_grad=True
print(x)
print(x.requires_grad)
```

```python
x=torch.tensor([1.],requires_grad=True)
y=x**2
print(y) # PowBackward0 가 붙어있다!

print(x.grad)
y.backward()
print(x.grad) # y=x**2을 미분한 2x의 x 값에 1을 대입한 gradient 값
# 이렇게 안하면, lim h->0 (x+h)-x / h 로 계산 한 부정확한 값이 나오겠징
```

```python
x=torch.tensor([1.], requires_grad=True)
y=x**2
print(y)
y.backward()
x.grad
# y.retain_grad() # 이걸 하면 y.grad도 볼 수 있다

# z=3*y
# print(z) # MulBackward0 가 붙어있다!

# z.backward()
# print(x.grad) # chain rule로 알아냄
# # print(y.grad) # warning! 중간건 안된다
```

```python
x=torch.tensor([2.] ,requires_grad=True)
x=x.detach() # detach도 많이 쓰이는 듯
# 윗 셀과 차이점은 데이터는 공유하지만 requires_grad는 다르게 지정이 가능하다는 점
print(x)
y=x**2
print(y)
```

```python
# detach 와 torch.no_grad
x=torch.tensor([1.], requires_grad=True)
# chain rule을 위해 계속 grad_fn을 update 하니까 grad_fn 잠시 안 계산하고 싶을 때 torch.no_grad
# 모델 테스트 시에는 불필요하게 메모리 쓸 필요 없기 때문! -> test / validation 할때 사용!
with torch.no_grad():
    y = x**2
    print(x.requires_grad)
    print(y) # with 안에서 계산되는 애는 grad_fn 이 안붙음
print(x.requires_grad)
# y.backward() # error!
y=x**2
print(y)

x=torch.tensor([1.], requires_grad=True)
x=x.detach()
y = x**2
print(x.requires_grad)
print(y)
# y.backward() # error!
```

## 간단한 인공신경망 만들기

```python
import torch
from torch import nn

x=torch.tensor([1.])
model=nn.Linear(1,1) # 입력 node 한 개, 출력 node 한 개인 layer 만듦
print(model)

print(model.weight) # 만들면서 initialize 함
print(model.bias)

y=model(x)
print(y)

y = x @ model.weight + model.bias
print(y)
```

```python
fc1=nn.Linear(1,3) # fully-connected
fc2=nn.Linear(3,1)

print(fc1.weight) # 3개 의 값 출력
print(fc1.bias) # 3개의 값 출력
print(fc2.weight) # 1개의 값 출력
print(fc2.bias) # 1개의 값 출력

x=torch.tensor([1.])
x=fc1(x)
print(x)
x=fc2(x)
print(x)

x = torch.tensor([1.])
y = (x@fc1.weight.T + fc1.bias) @ fc2.weight.T + fc2.bias
print(y)

# input size: 1
# fc1.weight.T size: 1x3
# fc1.bias size: 3
# fc2.weight.T size: 3x1
# fc2.bias size: 1
```

```python
model = nn.Linear(2,3)
x = torch.randn(10, 2 )
print(model(x))
# nn.Linear는 "채" 형태로(1D data) 들어오길 기대하는 녀석이다
# 즉, 노드 하나가 곧 한 채널을 의미한다.
# 따라서, 데이터 여러개를 통과시키고 싶다면 개x채 의 형태로 줘야 함

# why T? weight 도 데이터와 마찬가지로 input 개수x채널 형태로 만들기 위함!
# 예를 들어, nn.Linear(2,3) 이면 두 채널 값을 가지고 3개의 노드를 만드는 것이라 3x2 가 된다!
# 데이터의 개채는 두 채널 값 가지는 열개 데이터 (10x2)
# 웨이트의 개채는 두 채널 값으로 세 개 노드 만들기 (3x2)
```

```python
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(2,5)
        self.fc2 = nn.Linear(5,10)
        self.fc3 = nn.Linear(10,3)
        self.act = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.act(x)
        x = self.fc2(x)
        x = self.act(x)
        x = self.fc3(x)
        x = self.act(x)
        return x

model = MyModel()
x = torch.randn(5,2)
y = model(x) # model.forward(x)
print(y)
```
